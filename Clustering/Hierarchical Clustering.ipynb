```python
#Import libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Universities.csv')
df.head()
```

```python
x = df.drop(columns = 'Univ')
x
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
x = sc.fit_transform(x)
x
```

```python
from scipy.cluster import hierarchy
```

```python
lk = hierarchy.linkage(x)
dendrogram = hierarchy.dendrogram(lk)
```

```python
lk = hierarchy.linkage(x, method = 'complete')
dendrogram = hierarchy.dendrogram(lk)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters = 4)
y = hc.fit_predict(x)
```

```python
y
```

```python
df['Clusters'] = y
df
```

```python
df.groupby('Clusters').agg('mean')
```

```python
df[df['Clusters'] == 2]
```

```python
df[df['Clusters'] == 1]
```

```python
sns.scatterplot(df['Expenses'], df['Clusters'])
```

```python
sns.scatterplot(df['SAT'], df['Expenses'], hue = df['Univ'])
plt.legend(bbox_to_anchor = (1,1))
```

```python
sns.scatterplot(df['Clusters'], df['Expenses'], hue = df['Univ'])
plt.legend(bbox_to_anchor = (1,1))
```
